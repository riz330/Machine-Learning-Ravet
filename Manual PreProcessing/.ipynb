```python
import pandas as pd
df=pd.read_csv("Cars93.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
m=df.isna().sum()
```

```python
m[m>0]
```

```python
X=df.drop(columns=["id","Weight"])
```

```python
Y=df["Weight"]
```

```python
cat=X.columns[X.dtypes=="object"]
```

```python
cat
```

```python
con=X.columns[X.dtypes!="object"]
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
```

```python
num_pipe1=make_pipeline(SimpleImputer(strategy="mean"),StandardScaler())
```

```python
cat_pipe1=make_pipeline(SimpleImputer(strategy="most_frequent"),OrdinalEncoder())
```

```python
from sklearn.compose import ColumnTransformer
```

```python
pre1=ColumnTransformer(
    [
     ("num",num_pipe1,con),
     ("cat",cat_pipe1,cat)

    ]
).set_output(transform="pandas")
```

```python
pre_sel=pre1.fit_transform(X)
```

```python
pre_sel.head()
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
```

```python
sel_col=SequentialFeatureSelector(model1,n_features_to_select="auto",direction="forward")
```

```python
sel_col.fit_transform(pre_sel,Y)
```

```python
selected_column=sel_col.get_feature_names_out()
```

```python
selected_column
```

```python
sel_imp_col=[col.split("__")[1] for col in selected_column]
```

```python
sel_imp_col
```

```python
data=df[sel_imp_col]
```

```python
data.head()
```

```python
cat_sel=data.columns[data.dtypes=="object"]
```

```python
con_sel=data.columns[data.dtypes!="object"]
```

```python
cat_sel
```

```python
num_pipe2=make_pipeline(SimpleImputer(strategy="mean"),StandardScaler())
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_pipe2=make_pipeline(SimpleImputer(strategy="most_frequent"),OneHotEncoder(handle_unknown="ignore",sparse_output=False))
```

```python
pre2=ColumnTransformer(
    [
     ("num",num_pipe2,con_sel),
     ("cat",cat_pipe2,cat_sel)   
    ]
).set_output(transform="pandas")
```

```python
sel_pre=pre2.fit_transform(data,Y)
```

```python
sel_pre
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(sel_pre,Y,random_state=21,test_size=0.2)
```

```python
final_model=LinearRegression()
```

```python
final_model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.4f}")
    print(f"R2 score : {r2:.4f}")
```

```python
evaluate_model(final_model,xtrain,ytrain)
```

```python
evaluate_model(final_model,xtest,ytest)
```

```python
xnew_pre=pre2.transform(xnew)
```

```python
xnew_pre
```

```python
pred=final_model.predict(xnew_pre)
```
