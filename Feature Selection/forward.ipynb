### Step 4- Forward Feature Selection
one by one column is selected to get highest r2 score

```python
cat=list(X.columns[X.dtypes=="object"])
```

```python
cat
```

```python
con=list(X.columns[X.dtypes!="object"])
```

```python
con
```

```python
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1=make_pipeline(SimpleImputer(strategy="mean"),StandardScaler())
```

```python
cat_pipe1=make_pipeline(SimpleImputer(strategy="most_frequent"),OrdinalEncoder())
```

```python
pre1=ColumnTransformer(
    [
        ("num",num_pipe1,con)
        ,("con",cat_pipe1,cat)
    ]
).set_output(transform="pandas")
```

```python
X_pre=pre1.fit_transform(X)
```

```python
X_pre.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
model1=LinearRegression()
sel1=SequentialFeatureSelector(model1,n_features_to_select="auto",direction="forward")
sel1.fit_transform(X_pre,Y)
sel_cols=sel1.get_feature_names_out()
```

```python
sel_cols
```

```python
sel_cols[0].split("__")[1]
```

```python
imps_col=[col.split("__")[1] for col in sel_cols]
```

```python
imps_col
```

```python
X_sel=df[imps_col]
```

```python
X_sel.head()
```

```python
cat_sel=list(X_sel.columns[X_sel.dtypes=="object"])
con_sel=list(X_sel.columns[X_sel.dtypes!="object"])
```

```python
cat_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2=make_pipeline(SimpleImputer(strategy="mean"),StandardScaler())
```

```python
cat_pipe2=make_pipeline(SimpleImputer(strategy="most_frequent"),OneHotEncoder(handle_unknown="ignore",sparse_output=False))
```

```python
pre2 = ColumnTransformer(
    [
        ("num", num_pipe2, con_sel),
        ("cat", cat_pipe2, cat_sel)
    ]
).set_output(transform="pandas")
```

```python
X_sel_pre=pre2.fit_transform(X_sel)
```

```python
X_sel_pre.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X_sel_pre,Y,test_size=0.2,random_state=21)
```

```python
final_model=LinearRegression(
)
```

```python
final_model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.4f}")
    print(f"R2 score : {r2:.4f}")
```

```python
evaluate_model(final_model,xtrain,ytrain)
```

```python
evaluate_model(final_model,xtest,ytest)
```
