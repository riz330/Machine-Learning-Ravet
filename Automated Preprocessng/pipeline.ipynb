### Creating the Numerical Pipleine

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe=make_pipeline(SimpleImputer(strategy="mean"),StandardScaler())
```

```python
cat_pipe=make_pipeline(SimpleImputer(strategy="most_frequent"),OneHotEncoder(handle_unknown="ignore",sparse_output=False))
```

```python
# Combine both with column transformer
```

```python
pre=ColumnTransformer(
    [
        ("num",num_pipe,con),
        ("cat",cat_pipe,cat)
    ]


).set_output(transform="pandas")
```

```python
x_pre=pre.fit_transform(X)
```

```python
x_pre.head()
```

### Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_pre,Y,test_size=0.2,random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
from sklearn.metrics import(root_mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,r2_score)
```

```python
def evaluate(model,x,y):
    ypred=model.predict(x)
    #Evaluate
    rmse=root_mean_squared_error(y,ypred)
    mae=mean_absolute_error(y,ypred)
    mape=mean_absolute_percentage_error(y,ypred)
    r2=r2_score(y,ypred)
    #print
    print(f"root_mean_squared_error :{rmse}")
    print(f"mean absoulte error : {mae}")
    print(f"mean absolute percentage error :{mape}")
    print(f"r2 score :{r2}")
```

```python
evaluate(model,xtrain,ytrain)
```

```python
evaluate(model,xtest,ytest)
```
