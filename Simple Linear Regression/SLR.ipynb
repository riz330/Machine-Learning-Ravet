# Simple Linear Regression

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

## Step 1 - Read the Dataset

```python
import pandas as pd
df=pd.read_csv("50_Startups.csv")
df.head()
```

## I want to understand relationship between Profit And RND
## RND ~ Profit

# Profit_pred=B0+B1*RND

X= RND
Y= Profit

```python
X=df[["RND"]]
Y=df[["PROFIT"]]
```

```python
import seaborn as sns
sns.scatterplot(df,x="RND",y="PROFIT")
```

# Step 3 - Perform Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X,Y)
```

# Get Slope And Intercept

```python
model.intercept_
```

```python
model.coef_
```

# Profit_pred=49032 + 0.8542*RND

## Regression Plot

```python
sns.regplot(df,x="RND",y="PROFIT",line_kws={"color":"red"})
```

```python
sns.pairplot(df)
```

# Model Evaluation
1. Root Mean Square Error
2. Mean Absolute Error
3. R2 Score

```python
from sklearn.metrics import (root_mean_squared_error,mean_absolute_error,r2_score,mean_absolute_percentage_error)
```

```python
def evaluate(model,x,y):
    # Predict the results for x
    ypred=model.predict(x)
    # Evaluate the metrics
    rmse=root_mean_squared_error(y,ypred)
    mae=mean_absolute_percentage_error(y,ypred)
    r2=r2_score(y,ypred)
    mape=mean_absolute_percentage_error(y,ypred)
    # print the metrics
    print(f"RMSE :{rmse:.2f}")
    print(f"mean_absolute_error :{mae:.2f}")
    print(f"mean_absolute_percentage_error :{mape:.2f}")
    print(f"r2_score : {r2:.2f}")
```

```python
evaluate(model,X,Y)
```

From above r2 score is 0.95 or 95% which is greater than 80 which means this model is reliable
